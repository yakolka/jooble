```python
def preprocessing(df, factor, norm_type):
    
#I run follow preprocessing in case factor == 2, in others cases will create different "if" statements 
    if factor == 2:       
        
    
        df0 = df['id_job']
#extracting values from features column
        df = df['features'].str.split(",",expand=True)


        df = df.astype(float)

        df1 = df.iloc[:, 0]
        df = df.iloc[:, 1:]

#count columns max_feature_2_index and max_feature_2_abs_mean_diff
        df["max_feature_2_index"] = df.idxmax(axis=1)
        df["max_feature_2_abs_mean_diff"] = abs(df.max(axis=1) - (df.max(axis=1)).mean())


        df2 = df.iloc[:, 256:]
        df = df.iloc[:, :256]
#I run follow preprocessing in case normalization type is  z-type, in others cases will create different "if" statements         
        if norm_type == "z-type":
#normalization columns
            df = (df - df.mean())/df.std()
#creating new for names for columns
            df.columns = ['feature_2_' + str(col)  for col in df.columns]

            result = pd.concat([df0, df1, df, df2], axis=1, sort=False)
            result.columns.values[1] = "code"
            result.to_csv('C:/Users/Iryna/Desktop/jooble/result.tsv', sep = '\t')
            return result
        else:
            return 0
    else:
        return 0
```

```python
preprocessing(df, 2, "z-type")
```
